```r
### Pacotes necessários
#install.packages('arules', dep=T)
library(arules)
library(datasets)
```

```r
dados <- read.csv("databases/2 - Musculacao - Dados.csv", header = FALSE)

head(dados)
summary(dados)

set.seed(2034)

rules <- apriori(dados, parameter = list(supp = 0.001, conf = 0.7, minlen = 2))
summary(rules)


options(digits = 2)

inspect(sort(rules, by = "confidence"))
```
